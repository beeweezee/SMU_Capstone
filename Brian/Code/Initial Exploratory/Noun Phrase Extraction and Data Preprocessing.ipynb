```python
#some of the code below was borrowed from: https://www.kaggle.com/sudosharma/quick-and-dirty-entity-extractions


import spacy
import numpy as np
import text_normalizer as tn
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
file_info = "C:\\Users\\blgai\\OneDrive\\Documents\\School\\SMU\\Courses\\Fall 2021\\Capstone A\Data\\small_sampling_acc_data.csv"
df = pd.read_csv(file_info, sep='\t')
#concatenate title, summary, key_words and content columns so that we have more data to work with
df['combined']=df.apply(lambda x:'%s %s %s %s' % (x['title'],x['summary'],x['key_words'],x['content']),axis=1)
df.head()
```

```python
#now, we'll look for named entities and create a new column for that
nlp = spacy.load('en_core_web_sm')
tn.add_nouns(df)
df.head()
```

```python
tn.add_compounds(df)
df.head()
```

```python
#combine nouns and compound nouns
cols = ['nouns','compounds']
tn.add_comp_nouns(df,cols=cols)
df.head()
```

```python
def convert_to_string(text):
    s = str(text)
    return s

df['flat_comp_nouns'] = df.apply(lambda x: convert_to_string(x['comp_nouns']),axis=1)
```

```python
import nltk
stopword_list = nltk.corpus.stopwords.words('english')
#keep negation if in any bi-grams
stopword_list.remove('no')
stopword_list.remove('not')

#normalize the corpus
norm_corpus = tn.normalize_corpus(corpus=df['flat_comp_nouns'],
                                  html_stripping=True,
                                  contraction_expansion=True,
                                  text_lemmatization=True,
                                  text_stemming=False,
                                  special_char_removal=True,
                                  remove_digits=True,
                                  stopword_removal=True,
                                  stopwords=stopword_list
                                 )
df['clean_comp_nouns'] = norm_corpus
```
