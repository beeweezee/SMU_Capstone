```python
import pandas as pd
```

```python
file_info = "C:\\Users\\blgai\\OneDrive\\Documents\\School\\SMU\\Courses\\Fall 2021\\Capstone A\Data\\cleaned_combined_key_words_v1.csv"
df = pd.read_csv(file_info, sep='\t')
```

```python
df
```

```python
df.groupby(df.category).size()
```

```python
from sklearn.model_selection import train_test_split
#get 10% holdout set
train_corpus, holdout_corpus, train_label, holdout_label = train_test_split(df['clean_comp_nouns'], 
                                                                           df['category'],
                                                                           test_size=.10,random_state=1234)

#get 80% train and 20% test sets
train_corpus, test_corpus, train_label, test_label = train_test_split(train_corpus,
                                                                      train_label,
                                                                      test_size=.2225,random_state=1234) #.9 * .2225 = .20

train_corpus.shape, test_corpus.shape, holdout_corpus.shape
```

```python
df_train_comp_nouns = pd.DataFrame(train_corpus,columns=['clean_comp_nouns'])
df_train_comp_nouns['category'] = pd.DataFrame(train_label,columns=['category'])

df_holdout_comp_nouns = pd.DataFrame(holdout_corpus,columns=['clean_comp_nouns'])
df_holdout_comp_nouns['category'] = pd.DataFrame(holdout_label,columns=['category'])

df_test_comp_nouns = pd.DataFrame(test_corpus,columns=['clean_comp_nouns'])
df_test_comp_nouns['category'] = pd.DataFrame(test_label,columns=['category'])
```

```python
file_out = "C:\\Users\\blgai\\OneDrive\\Documents\\School\\SMU\\Courses\\Fall 2021\\Capstone A\Data\\train_comp_nouns_v1.csv"
df_train_comp_nouns.to_csv(file_out,index=False)

file_out = "C:\\Users\\blgai\\OneDrive\\Documents\\School\\SMU\\Courses\\Fall 2021\\Capstone A\Data\\test_comp_nouns_v1.csv"
df_test_comp_nouns.to_csv(file_out,index=False)

file_out = "C:\\Users\\blgai\\OneDrive\\Documents\\School\\SMU\\Courses\\Fall 2021\\Capstone A\Data\\holdout_comp_nouns_v1.csv"
df_holdout_comp_nouns.to_csv(file_out,index=False)
```

```python
from sklearn.model_selection import train_test_split
#get 10% holdout set
train_corpus, holdout_corpus, train_label, holdout_label = train_test_split(df['clean_key_words'], 
                                                                           df['category'],
                                                                           test_size=.10,random_state=1234)

#get 80% train and 20% test sets
train_corpus, test_corpus, train_label, test_label = train_test_split(train_corpus,
                                                                      train_label,
                                                                      test_size=.2225,random_state=1234) #.9 * .2225 = .20

train_corpus.shape, test_corpus.shape, holdout_corpus.shape
```

```python
df_train_keywords = pd.DataFrame(train_corpus,columns=['clean_key_words'])
df_train_keywords['category'] = pd.DataFrame(train_label,columns=['category'])

df_holdout_keywords = pd.DataFrame(holdout_corpus,columns=['clean_key_words'])
df_holdout_keywords['category'] = pd.DataFrame(holdout_label,columns=['category'])

df_test_keywords = pd.DataFrame(test_corpus,columns=['clean_key_words'])
df_test_keywords['category'] = pd.DataFrame(test_label,columns=['category'])
```

```python
file_out = "C:\\Users\\blgai\\OneDrive\\Documents\\School\\SMU\\Courses\\Fall 2021\\Capstone A\Data\\train_keywords_v1.csv"
df_train_keywords.to_csv(file_out,index=False)

file_out = "C:\\Users\\blgai\\OneDrive\\Documents\\School\\SMU\\Courses\\Fall 2021\\Capstone A\Data\\test_keywords_v1.csv"
df_test_keywords.to_csv(file_out,index=False)

file_out = "C:\\Users\\blgai\\OneDrive\\Documents\\School\\SMU\\Courses\\Fall 2021\\Capstone A\Data\\holdout_keywords_v1.csv"
df_holdout_keywords.to_csv(file_out,index=False)
```
