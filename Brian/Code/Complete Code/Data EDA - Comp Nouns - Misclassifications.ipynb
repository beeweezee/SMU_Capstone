```python
import nltk
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import warnings
import model_report as mr
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
file_info = "C:\\Users\\blgai\\OneDrive\\Documents\\School\\SMU\\Courses\\Fall 2021\\Capstone A\Data\\ensemble_misclassifications_v1.csv"
df = pd.read_csv(file_info)
```

```python
df.head()
```

```python
df['category'].unique()
```

```python
from collections import Counter
df_group_by = df.groupby('category')['terms'].apply(lambda x: Counter(' '.join(x).split()).most_common(25))
```

```python
df_group_by[6]
```

# Frequency of 25 Most Common Words

```python
#Business
group_by_bu = df_group_by.iloc[0]
words_bu = list(zip(*group_by_bu))[0]
freq_bu = list(zip(*group_by_bu))[1]

plt.figure(figsize=(50,30))
plt.bar(words_bu, freq_bu)
plt.xlabel('Words', fontsize=50)
plt.ylabel('Frequency of Words', fontsize=50)
plt.yticks(fontsize=40)
plt.xticks(rotation=60, fontsize=40)
plt.title('Frequency of 25 Most Common Words for Business', fontsize=60)
plt.show()
```
